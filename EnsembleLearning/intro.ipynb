# Ensemble Learning 

- Ensemble : A group of artists

- It is a collection of machine learning models , combine them and made a model.This technique is ensemble learning

- Wisdom of the Crowd

    - Real Life Examples:
       
       - Audience Poll KBC , maximum voted is mostly correct.

       - User Reviews , then maximum is mostly true

Core Idea : How to create a difference 

1. The models are different with training on same data.

2. The models are with training on diff data.

3. The models and training data both are different.

# Type of Ensemble Learning

- Voting Ensemble

    - Here Based Models are different algorithms , providing the same data.
    
    - Then take the majority count in case of classification.

    - Take the mean in case of regression.

    - Since diff models , so there is variety.

      

- Stacking
    
    - Same as voting but in voting all the models are considered equal or have equal weightage , but in this stacking 
      some models whose results are more accuracte give me more weightage.

    - The final model is traing on the prediction of others and then assigning weightage to them during training.

      

- Bagging 

    - Bagging stands for Boostraped aggregation .
    - In this multilple models (typically of same type) are trained on diff data , and at time of testing their results are aggregated.
    - If models used is Decision Tree , then we called it a RandomForest.

      


- Boosting 

    - In this first model trained , it note down its mistakes and then it will tell the mistakes to next model.
      Then we see improved results on next one , slly it will send to next .
    - In this way , whole model will improve.
    - Some of the most Famous Algorithms used : AdaBoost , Gradient Boosting , XGBoost
