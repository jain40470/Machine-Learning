Measure of Central Tendency 

  - It is a statistical measure that represents a typical or central value for a dataset. Eg : For a column like , you are finding a central value 

  - Here we calculate  
    
    - Mean (Sum of all / no of values)

    - Median (Middle when arranged in order)

    - Mode (Most occured value)

    - Weighted Mean : (sum of pdt of value and wt / sum of weights of value) , It is used when values in dataset have diff importance or frequency.

    - Trimmed Mean : Remove some small and large values , then calculating the mean.

Measure of Dispersion

- It is a statistical measure that describes the spread and variability of a dataset.
- It provides information about how data is  distributed about the central tendency (mean , median , mode) of a dataset.
- Some measure of dispersion 

  - Range : Diff b/w maximum and minimum value , simple but affected by outliers.

  - Variance : It is the average of squared diff b/w each data point and mean. It measures the average distance of each data point from mean. It is not exactly spread but propotional to spread.
    Mean Absolute devaition : same as Variance , here instead of square , we are doing absolute , but we cannot have inference from it. (if we know sample MAD , then we cannot get population MAD)
    - Population Variance : [ summan : (x - u)^2  / N  ] , where u is mean and N is total no of items in population.
    - Sample Variance : [ summan : (x - u)^2  / n-1  ] , where u is mean and n is total no of items in sample.
  
  - Standard Deviation : It is the square root of devaition. It's unit is same as data unit. (It also have similar population std. deviation and sample std. deviation)

  - Coefficient of Variation (CV) : (Std Deviation / Mean)*(100 %) , let say you have two column salary and experience.
Their mean and std deviation can be compared seperately as they are completely diff .
So now if want to know in which column the data is more near to mean , we can , measure CV.


Boxplot

- A box plot, also known as a box-and-whisker plot, is a graphical representation of a dataset that shows the distribution of the data. The box plot displays a summary of the data, including the minimum and maximum values, the first quartile (Q1), the median (Q2), and the third quartile (Q3).



- Calculation while plotting box plot

  - Q1 : (25/100) * (N + 1)  (no at this , if not exact take average of upper and lower)
  - Q2 : (50/100) * (N + 1) 
  - Q3 : (75/100) * (N + 1)
  - IQR : Q3-Q1
  - Min : Q1 - 1.5 * IQR
  - Min : Q3 + 1.5 * IQR

  


- Benefits 

  - Easy to see the distribution of data.
  
  - Can identify outliers.

  - Tells about the skewness of data.

    

  - Compare 2 categories of data.

    



```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


gender = np.random.choice(['Male', 'Female'], 100)
weight = np.random.normal(5, 1, 100)  # Average weight 5 kg

df = pd.DataFrame({'Gender': gender, 'Weight': weight})

sns.boxplot(x='Gender', y='Weight', data=df)
plt.title('Weight vs Gender (Cats)')
plt.show()
```
