Measure of Central Tendency 

  - It is a statistical measure that represents a typical or central value for a dataset. Eg : For a column like , you are finding a central value 

  - Here we calculate  
    
    - Mean (Sum of all / no of values)

    - Median (Middle when arranged in order)

    - Mode (Most occured value)

    - Weighted Mean : (sum of pdt of value and wt / sum of weights of value) , It is used when values in dataset have diff importance or frequency.

    - Trimmed Mean : Remove some small and large values , then calculating the mean.

Measure of Dispersion

- It is a statistical measure that describes the spread and variability of a dataset.
- It provides information about how data is  distributed about the central tendency (mean , median , mode) of a dataset.
- Some measure of dispersion 

  - Range : Diff b/w maximum and minimum value , simple but affected by outliers.

  - Variance : It is the average of squared diff b/w each data point and mean. It measures the average distance of each data point from mean. It is not exactly spread but propotional to spread.
    Mean Absolute devaition : same as Variance , here instead of square , we are doing absolute , but we cannot have inference from it. (if we know sample MAD , then we cannot get population MAD)

    - Population Variance : [ summan : (x - u)^2  / N  ] , where u is mean and N is total no of items in population.
    - Sample Variance : [ summan : (x - u)^2  / n-1  ] , where u is mean and n is total no of items in sample.
  
       
    
  - Standard Deviation : It is the square root of devaition. It's unit is same as data unit. (It also have similar population std. deviation and sample std. deviation)

  - Coefficient of Variation (CV) : (Std Deviation / Mean)*(100 %) , let say you have two column salary and experience.
Their mean and std deviation can be compared seperately as they are completely diff .
So now if want to know in which column the data is more near to mean , we can , measure CV.


# Boxplot

- A box plot, also known as a box-and-whisker plot, is a graphical representation of a dataset that shows the distribution of the data. The box plot displays a summary of the data, including the minimum and maximum values, the first quartile (Q1), the median (Q2), and the third quartile (Q3).

  

- Calculation while plotting box plot

  - Q1 : (25/100) * (N + 1)  (no at this , if not exact take average of upper and lower)
  - Q2 : (50/100) * (N + 1) 
  - Q3 : (75/100) * (N + 1)
  - IQR : Q3-Q1
  - Min : Q1 - 1.5 * IQR
  - Min : Q3 + 1.5 * IQR

  


- Benefits 

  - Easy to see the distribution of data.
  
  - Can identify outliers.

  - Tells about the skewness of data.

    

  - Compare 2 categories of data.

    


How to Tell Skewness:

Skewness Type	

- Symmetric	
    - Median is in the center of the boxWhiskers are approximately equal
- Right-Skewed (Positive)	
    - Median is closer to Q1 (left side of box)
    - Right whisker is longer
    - More outliers on the right
- Left-Skewed (Negative)	
    - Median is closer to Q3 (right side of box)
    - Left whisker is longer
    - More outliers on the left

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
gender = np.random.choice(['Male', 'Female'], 100)
weight = np.random.normal(5, 1, 100)  # (mean, std_dev, size)
df = pd.DataFrame({'Gender': gender, 'Weight': weight})
```

```python
sns.boxplot(y='Weight', data=df)
plt.title('Box Plot of Cat Weights')
plt.show()
```

```python

sns.boxplot(x='Gender', y='Weight', data=df)
plt.title('Weight vs Gender (Cats)')
plt.show()
```

# Covariance



   - Why Covariance exist , What does Covariance solve ?

      - Mean just tell us about the center not the about spread , for that we have variance.

      - Let consider these two ( (-1,-1) , (0,0) , (1,1) ) and ( (-1,1) , (0,0) , (1,-1)) , you will get same variance.
        So according to variance , these two are same.

      - To distinguish b/w these we have covariance.

        

      - It gives nature of linear relationship. 




   - What is Covariance and How is it interpreted ? 

       - Covariance is a statistical measure that describes the degree to which two variables are linearly related. It measures how much two variables change together, such that when one variable increases, does the other variable also increase, or does it decrease?

       - If the covariance between two variables is positive, it means that the variables tend to move together in the same direction. If the covariance is negative, it means that the variables tend to move in opposite directions. A covariance of zero indicates that the variables are not linearly related.

       - Let say we plot a scatterplot of x and y.

          

       - How is it calculated 

          



   - Limitation of Covariance 
   
      - One limitation of covariance is that it does not tells us about the strength of relationship between 
        two variables , since the magnitude of covariance is affected by scale of the variables.

        

      - Like in figure 1 : they are more related and in 2nd it is spread but nature of linear relationship is same
        in both.


Representation of How magnitude of covariance change by scaling.
Shows that the magnitude of covariance is significant.

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame()
x = pd.Series( [12,25,68,42,113] )
y = pd.Series( [11,29,58,121,100] )
df['x'] = x
df['y'] = y
df
```

```python
fig , ax = plt.subplots( 1 , 3 ,figsize=(15,4))
# 1, 3: This means you are creating 1 row and 3 columns of subplots
# figsize=(15, 3): Sets the overall size of the figure to 15 units wide and 4 units tall. 

ax[0].scatter(df['x'] , df['x'])
ax[1].scatter(df['x'] , df['y'])
ax[2].scatter(df['x']*2 , df['y']*2)

ax[0].set_title( "Covariance - " + str( np.cov(df['x'],df['x'])[0,1] ) )
ax[1].set_title( "Covariance - " + str( np.cov(df['x'],df['y'])[0,1] ) )
ax[2].set_title( "Covariance - " + str( np.cov(df['x']*2,df['y']*2)[0,1] ) )
```

# Correlation

- Correlation refers to a statistical relationship between two or more variables. Specifically , it measures the degree 
to which the two variables are related and how they tend to change together.

- It is in range of -1 to 1.

- Correlation is often measured using a statistical tool called the correlation coefficient which ranges from 
  -1 to 1 . A correlation coefficient of -1 indicates , a perfect negative correlation , a correlation 
  coefficient 0 indicated no correlation , and a correlation of 1 indicates a perfect positive correlation.

-   

-  Formula is Covariance(x,y) / (std devn of x * std devn of y)
    

- This correlation not affected by scaling.

# Correlation and Causation

- The phrase "correlation does not imply causation" means that just because two variables are associated with each other, it does not necessarily mean that one causes the other.

- In other words, a correlation between two variables does not necessarily imply that one variable is the reason for the other variable's behaviour.

- Example : 
  - Suppose there is a positive correlation between the number of firefighters present at a fire and the amount of damage caused by the fire. 
  - One might be tempted to conclude that the presence of firefighters causes more damage.
  -  However, this correlation could be explained by a third variable - the severity of the fire. More severe fires  might require more firefighters to be present, and also cause more damage.

- Thus, while correlations can provide valuable insights into how different variables are related, they cannot be used to establish causalfty. 

- Establishing causality often requires additional evidence such as experiments, randomized controlled trials, or well-designed observational studies.
