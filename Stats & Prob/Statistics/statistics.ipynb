# Statistics 

Involves collecting , analysing , interpreting and presenting data.

Population : Refers to the entire group of individuals or objects that we are interested in studying.It is the complete 
set of observations that we want to make inference about.

Sample : It is a subset of population , that we select from population to study.

Things to take care while creating samples
- Sample size
- Random
- Representative

A parameter is a characterstic of population , while statistic is a characterstic of a sample.

Two Types :

1. Descriptive Statistics
    
   Involves generation of summary and describing features of a set of data. [Like study of data]

2. Inferential Statistics
  
   Involves predicting results and conclusion out of data.
